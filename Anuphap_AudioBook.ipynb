# Create an Audiobook from a PDF
## This task tests your ability to apply Text to Speech conversion and Extraction of Text from PDF files in the creation of an audiobook from a PDF file

### Steps
- Extract text from PDF file
- Clean the text
- Convert the text into speech
- Save the speech
- Play the speech

## 1. Extract text from PDF
- Use PyPDF2

### Install the library

```python
# ! pip install PyPDF2
```

### Import the library

```python
import PyPDF2
```

### Extract the text

```python
def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        num_pages = len(reader.pages)
        text = ''
        
        for page_num in range(num_pages):
            page = reader.pages[page_num]
            text += page.extract_text()
        
        return text

pdf_file_path = 'The-Time-Machine.pdf' #Link Ref:https://manybooks.net/books/get/127836/6
```

### Print the extracted text

```python
def print_extracted_text(pdf_path):
    extracted_text = extract_text_from_pdf(pdf_path)
    print(extracted_text)
    
print_extracted_text(pdf_file_path)
extracted_text = extract_text_from_pdf(pdf_file_path)
```

## 2. Convert the Text into Speech
- Use **pyttsx3** OR **gTTS**

### Install the library

```python
# !pip install pyttsx3
```

### Import the library

```python
from gtts import gTTS
```

### Initialize a Speaker object

```python
text='Hi ,Welcome to GBC AI Program'
```

```python
language = 'en'
```

```python
speech = gTTS(text = text, lang = language, slow = False)
```

```python
speech.save('speech_1.wav')
```

### Convert the text

```python
speech = gTTS(text=extracted_text, lang='en')
```

### Save the audio

```python
speech.save('Audiobook_The_Time_Machine.wav')
```
